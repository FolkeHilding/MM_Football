```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch, Sbopen
```

```python
parser = Sbopen()
df, related, freeze, tactics = parser.event(69301)
passes = df.loc[df['type_name'] == 'Pass'].loc[df['sub_type_name'] != 'Throw-in'].set_index('id')
```

```python
sweden_passes = passes.loc[passes['team_name'] == 'Sweden']
print(f"Number of Sweden passes: {len(sweden_passes)}")
print(sweden_passes[['x', 'y', 'end_x', 'end_y']].describe())
pitch = Pitch(line_color = "black")
fig, ax = pitch.draw(figsize=(10, 7))
pitch.scatter(sweden_passes['x'], sweden_passes['y'], ax=ax, color='blue', edgecolors='black', label='Sweden Passes')
```
